```python
#najprv si naimportujem knižnice, ktoré budem používať
import tensorflow
import pandas as pd
import seaborn as sns
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#pomocou knižnice pandas si načítame csv súbor
df = pd.read_csv('housepricedata.csv')
```

```python
resizeplot(10,5)
sns.countplot(x='TotRmsAbvGrd',data=df)  #príklad ako by sme si vykreslili  informácie o datach pomocou grafu(konkrétne počet izieb)
```

```python
#Množina údajov, ktorú teraz máme, sa nachádza v dátovom rámci pandas, prevedieme ho na pole aby sme jednoducho získali hodnoty
dataset = df.values
dataset
```

```python
#dataset sme  rozdelili na vstupné funkcie (X) a funkciu, ktorú chceme predpovedať (Y)
X = dataset[:,0:10]
Y = dataset[:,10]
```

```python
#normalizácia dát je veľmi dôležitá, použijem min-max scaler od scikit-learn, ktorý upraví dáta tak, aby boli v rozmedzí 0 a 1.
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
```

```python
X_scale #zobrazím si x_scale
```

```python
#teraz si potrebujem rozdeliť dáta na trénovacie a testovacie, na to použijeme funckiu train_test_split z knižnice scikit-learn
X_train, X_val_and_test, Y_train, Y_val_and_test = train_test_split(X_scale, Y, test_size=0.3)
```

```python
#rozdelí veľkosť val_and_test rovnako na overovaciu sadu a testovaciu sadu
X_val, X_test, Y_val, Y_test = train_test_split(X_val_and_test, Y_val_and_test, test_size=0.5)
```

```python
#teraz máme celkovo šesť premenných pre naše množiny údajov, ktoré budeme používať
print(X_train.shape, X_val.shape, X_test.shape, Y_train.shape, Y_val.shape, Y_test.shape)
```

Trénovacia sada má 1022 dátových bodov, zatiaľ čo validačná a testovacia sada má po 219 dátových bodoch. Premenné X majú 10 vstupných funkcií, zatiaľ čo premenné Y majú iba jednu vlastnosť na predpovedanie.

```python
#definujem si architektúru siete
model = Sequential([   #model - sekvenčný
    Dense(32, activation='relu', input_shape=(10,)),   #hustota 32, vstupná vrstva 10
    Dense(32, activation='relu'),   #nemusíme popisovať vstupný tvar, pretože Keras ho môže odvodiť z výstupu našej prvej vrstvy
    Dense(1, activation='sigmoid'),  #tretia vrstva je hustá vrstva s 1 neurónom a aktiváciou sigmoidu
])
```

Link: https://keras.io/guides/sequential_model/

```python
model.compile(optimizer='sgd',  #Sgd označuje stochastický gradientový zostup
              loss='binary_crossentropy', #chybová funkcia pre výstupy, hodnoty 1 alebo 0
              metrics=['accuracy'])  #presnosť
```

```python
hist = model.fit(X_train, Y_train,  #určujem na ktorých údajoch trénujeme, teda X_train a Y_train
          batch_size=32, epochs=100,  #data si nasekam na tzv. "batches a určím počet iterácií
          validation_data=(X_val, Y_val)) #zadám validačné údaje
```

```python
model.evaluate(X_test, Y_test)[1] #vygenerujem si presnosť
```

```python
resizeplot(10,6) 
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('graf chybovosti')
plt.ylabel('chyby')
plt.xlabel('Počet iterácií')
plt.legend(['Train', 'Val'], loc='upper right')
plt.show()
```

```python
resizeplot(10,6)
plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])
plt.title('Model presnosti')
plt.ylabel('Presnosť')
plt.xlabel('Počet iterácií')
plt.legend(['Train', 'Val'], loc='lower right')
plt.show()
```

```python
from keras.layers import Dropout #importujem si kód, ktorý potrebujem na regularizáciu a dropout
from keras import regularizers
```

## Vyskúšam si regularizer
Váhová regularizácia umožňuje zníženie preťaženia modelu neurónovej siete na trénovacích dátach a zlepšuje výkonu modelu na testovacích.
Používa sa ako riešenie overfittingu. Vyskúšam si L2 regularizer
- L2 regularizer: Súčet váh na druhú.
Link: https://machinelearningmastery.com/how-to-reduce-overfitting-in-deep-learning-with-weight-regularization/

```python
model_3 = Sequential([ #regularizes - berie druhé mocniny parametrov do našej funkcie celkovej straty a zváži ich o 0,01 vo funkcii straty.
    Dense(1000, activation='relu', kernel_regularizer=regularizers.l2(0.01), input_shape=(10,)),
    Dropout(0.3), #znamená, že neuróny v predchádzajúcej vrstve majú pravdepodobnosť 0,3 vypadnutia počas tréningu
    Dense(1000, activation='relu', kernel_regularizer=regularizers.l2(0.01)),
    Dropout(0.3),
    Dense(1000, activation='relu', kernel_regularizer=regularizers.l2(0.01)),
    Dropout(0.3),
    Dense(1000, activation='relu', kernel_regularizer=regularizers.l2(0.01)),
    Dropout(0.3),
    Dense(1, activation='sigmoid', kernel_regularizer=regularizers.l2(0.01)),
])
model_3.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
hist_3 = model_3.fit(X_train, Y_train,
          batch_size=32, epochs=100,
          validation_data=(X_val, Y_val))
```

```python
resizeplot(10,6)
plt.plot(hist_3.history['loss'], label='Train set loss values')
plt.plot(hist_3.history['val_loss'], label='Validation set loss values')
plt.ylabel('Chyby')
plt.xlabel('Iterácie')
plt.legend(loc='upper right')
plt.show()
```
